```python
# Link dataset: https://www.kaggle.com/uciml/german-credit
import pandas as pd
import seaborn as sns
```

```python
german_df = pd.read_csv('german_credit_data.csv', index_col = 0)
german_df.head()
```

```python
german_df_1 = german_df.loc[(german_df['Age']>=20) & (german_df['Age']<=30)]
german_df_1.head()
```

```python
german_df_2 = german_df.loc[(german_df['Age']>=45) & (german_df['Age']<=60) & (german_df['Sex']=='female')]
german_df_2.head()
```

```python
german_df_2.to_csv('german_credit_data_new.csv')
```
